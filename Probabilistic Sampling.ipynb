```python
'''
Code by Sourabh Ghorpade
Description - Get the sampling of the data with the probability of each element getting selected is proportional to its value,
called probabilistic sampling.
'''
import numpy as np
import matplotlib.pyplot as plt
A = [0,5,27,6,13,28,100,45,10,79] #Data 

def propotional_sampling(A):
    sum=0
    for i in range(0,len(A)): #get the sum of all elememts
        sum+=A[i]
    Normalised_A=[]
    for i in range(0,len(A)): #Get the data normalised by dividing each element by the total sum
        Normalised_A.append((A[i]/sum))
    #print(Normalised_A)
    
    Cumulative_A=[]
    Cumulative_A.append(0)
    
    for i in range(1,len(A)): #Get the cumulative sum of the normalised data.
        #This way bigger number will have bigger share between 0 to 1 and so its probability is high to appear in random uniform
        #distribution.
        Cumulative_A.append(Normalised_A[i]+Cumulative_A[i-1])
    #print(Cumulative_A)
    
    pick=np.random.uniform(0,1,1) #Pick number from uniform distribution.
    
    for i in range(0,len(A)): #Get the respective element from A from range of cumulative sum to which the random no. belongs.
        if(pick<Cumulative_A[i]):
            return A[i]
    #print(pick)
def sampling_based_on_magnitued():
    nos=[]
    for i in range(1,100):
        number = propotional_sampling(A)
        nos.append(number)
        print(number)
    plt.hist(nos)
    plt.show()
sampling_based_on_magnitued()
```
